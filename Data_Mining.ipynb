```python
import pandas as pd

df = pd.read_csv('/content/Consumer_Electronics_Sales_Data.csv')
```

```python
# Column names
df.columns
```

```python
# Drop first column (ProductID)
df = df.drop('ProductID', axis=1)
```

```python
# Summary
summary = df.describe()
print(summary)
```

```python
# Count missing values in each column
missing_values_count = df.isna().sum()
print(missing_values_count)

# No Missing values Found
```

```python
# Categorizing the 'CustomerAge' column
# Young: 15 to 30
# Middle-age: 31 to 50
# Old-age: 51 to 70

# Bins and corresponding labels
bins = [15, 30, 50, 70]
labels = ['Young', 'Middle-age', 'Old-age']

# Categorizing
df['CustomerAge'] = pd.cut(df['CustomerAge'], bins=bins, labels=labels, right=False)

df.head()
```

```python
# Categorizing the 'PurchaseFrequency' column
# Occasional: 1 to 5
# Regular: 6 to 15
# Premium: 16 to 20

# Bins and corresponding labels
bins = [1, 5, 15, 20]
labels = ['Occasional', 'Regular', 'Premium']

# Categorizing
df['PurchaseFrequency'] = pd.cut(df['PurchaseFrequency'], bins=bins, labels=labels, right=False)

df.head()
```

```python
# Categorizing the 'ProductPrice' column
# Low: 1 to 1000
# Medium: 1001 to 2000
# High: 2001 to 3000

# Bins and corresponding labels
bins = [1, 1000, 2000, 3000]
labels = ['Low', 'Medium', 'High']

# Categorizing
df['ProductPrice'] = pd.cut(df['ProductPrice'], bins=bins, labels=labels, right=False)

df.head()
```

```python
# Replace values in the 'CustomerGender' column
df['CustomerGender'] = df['CustomerGender'].replace({0: 'Male', 1: 'Female'})

df.head()
```

```python
# Replace values in the 'CustomerSatisfaction' column
df['CustomerSatisfaction'] = df['CustomerSatisfaction'].replace({1: 'Dissatisfied', 2: 'Somewhat Dissatisfied', 3: 'Neutral', 4: 'Satisfied', 5: 'Very Satisfied'})

df.head()
```

```python
# Replace values in the 'PurchaseIntent' column
df['PurchaseIntent'] = df['PurchaseIntent'].replace({0: 'No', 1: 'Yes'})

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Creating a LabelEncoder object
le = LabelEncoder()

# Applying label encoding to each categorical column
for column in ['ProductCategory', 'ProductBrand', 'ProductPrice', 'CustomerAge', 'CustomerGender', 'PurchaseFrequency', 'CustomerSatisfaction', 'PurchaseIntent']:
    df[column] = le.fit_transform(df[column])

# Defining features and target variable
X = df.drop('PurchaseIntent', axis=1) # Contains Input
y = df['PurchaseIntent']              # Contains Output

# Spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Testing 70-30 split
X_train.shape, X_test.shape
```

```python
import numpy as np

def gaussian_naive_bayes(X_train, y_train, X_test):
    # Calculating class priors
    classes, counts = np.unique(y_train, return_counts=True)
    priors = counts / len(y_train)

    # Calculating mean and standard deviation for each feature in each class
    means = {}
    stds = {}
    for cls in classes:
        cls_data = X_train[y_train == cls]
        means[cls] = np.mean(cls_data, axis=0)
        stds[cls] = np.std(cls_data, axis=0)

    # Calculating class conditional probabilities using Gaussian distribution
    probs = []
    for cls in classes:
        class_prob = np.sum(-0.5 * ((X_test - means[cls]) ** 2) / (stds[cls] ** 2)
                            - 0.5 * np.log(2 * np.pi * (stds[cls] ** 2)), axis=1)
        probs.append(class_prob + np.log(priors[cls]))

    # Predicting the class with the highest probability
    y_pred = classes[np.argmax(probs, axis=0)]

    return y_pred
```

```python
# Trainiing and predicting using Gaussian Naive Bayes function
y_pred = gaussian_naive_bayes(X_train, y_train, X_test)
```

```python
# Evaluating the accuracy
accuracy = np.mean(y_pred == y_test)
print(f'Accuracy: {accuracy*100:.2f}%')
```
